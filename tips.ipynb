```python
for col in ['total_bill']:
    sns.histplot(pf[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in ['tip','size']:
    sns.histplot(pf[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in ['day','time']:
    sns.countplot(data=pf, x=col)
    plt.title(f"Count of {col}")
    plt.show()
```

```python
for col in ['sex','smoker']:
    sns.countplot(data=pf, x=col)
    plt.title(f"Count of {col}")
    plt.show()
```

```python
sns.lineplot(data=pf, x='total_bill', y='tip')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
pf['tip_pct']=pf['tip']/pf['total_bill']*100
sns.histplot(pf['tip_pct'],kde=True)
plt.title('Distribution of Tip Percentage')
plt.show()
```

```python
sns.lineplot(data=pf, x='day', y='tip_pct', hue='time')
plt.title("Tip % by Day & Time")
plt.show()
```

```python
sns.pairplot(pf)
```

```python
sns.scatterplot(data=pf, x="total_bill", y="tip_pct")
plt.title("Total Bill vs Tip Percentage")
plt.show()
```

Do higher total bills result in higher tip percentages?
Tip amount increases with total bill.

But tip percentage does NOT increase; it stays around 15â€“20%.
Larger bills â†’ bigger tips in amount, but percentage stays stable

```python
sns.boxplot(data=pf, x="smoker", y="tip_pct")
plt.title("Tip %: Smoker vs Non-Smoker")
plt.show()

pf.groupby("smoker")["tip_pct"].mean()
```

Do smokers tip more or less?
Smokers give slightly lower tip % than non-smokers.

Non-smokers generally tip better

```python
sns.scatterplot(data=pf, x="size", y="tip")
plt.title("Group Size vs Tip")
plt.show()

pf.groupby("size")["tip"].mean()
```

Do larger groups give higher tips?Bigger groups give higher total tip amount.

But tip percentage (% of bill) stays almost the same.

Groups of 4â€“6 tip the highest.

```python
pf.groupby("day")[["total_bill","tip"]].sum().sort_values("total_bill", ascending=False)
```

```python
sns.barplot(data=pf, x="day", y="total_bill", estimator=sum)
plt.title("Total Bill by Day")
plt.show()

sns.barplot(data=pf, x="day", y="tip", estimator=sum)
plt.title("Total Tip by Day")
plt.show()
```

```python
sns.boxplot(data=pf, x="sex", y="tip_pct")
plt.title("Tip % by Gender")
plt.show()

pf.groupby("sex")["tip_pct"].mean()
```
